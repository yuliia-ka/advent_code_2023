```python
# Part Two

def one_cycle(input_lines):
    input_lines = [list(row) for row in input_lines]
    previous_rocks_pos = []

    while True:
        rocks_pos = rock_positions(input_lines)
        if rocks_pos == previous_rocks_pos:
            break
        
        for pos_of_rocks_row,pos_of_rocks_column in rocks_pos:
            if pos_of_rocks_row > 0:
                next_coor = (pos_of_rocks_row-1,pos_of_rocks_column)
                if input_lines[next_coor[0]][next_coor[1]]=='.':
                    input_lines[next_coor[0]][next_coor[1]]='O'
                    input_lines[pos_of_rocks_row][pos_of_rocks_column]='.'
                if input_lines[next_coor[0]][next_coor[1]]=='#':
                    pass
                else:
                    pass
            previous_rocks_pos = rocks_pos

    # print('done north')  
    input_lines
    ####################################################################   
        
    previous_rocks_pos = []

    while True:
        rocks_pos = rock_positions(input_lines)
        if rocks_pos == previous_rocks_pos:
            break
        
        for pos_of_rocks_row,pos_of_rocks_column in rocks_pos:
            if pos_of_rocks_column > 0:
                next_coor = (pos_of_rocks_row,pos_of_rocks_column-1)
                if input_lines[next_coor[0]][next_coor[1]]=='.':
                    input_lines[next_coor[0]][next_coor[1]]='O'
                    input_lines[pos_of_rocks_row][pos_of_rocks_column]='.'
                if input_lines[next_coor[0]][next_coor[1]]=='#':
                    pass
                else:
                    pass
            previous_rocks_pos = rocks_pos
    # print('done west') 
    ####################################################################

    previous_rocks_pos = []

    while True:
        rocks_pos = rock_positions(input_lines)
        if rocks_pos == previous_rocks_pos:
            break
        
        for pos_of_rocks_row,pos_of_rocks_column in rocks_pos:
            if pos_of_rocks_row < len(input_lines[0])-1:
                next_coor = (pos_of_rocks_row+1,pos_of_rocks_column)
                if input_lines[next_coor[0]][next_coor[1]]=='.':
                    input_lines[next_coor[0]][next_coor[1]]='O'
                    input_lines[pos_of_rocks_row][pos_of_rocks_column]='.'
                if input_lines[next_coor[0]][next_coor[1]]=='#':
                    pass
                else:
                    pass
            previous_rocks_pos = rocks_pos
    # print('done south')   

    ####################################################################
    previous_rocks_pos = []

    while True:
        rocks_pos = rock_positions(input_lines)
        if rocks_pos == previous_rocks_pos:
            break
        
        for pos_of_rocks_row,pos_of_rocks_column in rocks_pos:
            if pos_of_rocks_column < len(input_lines)-1:
                next_coor = (pos_of_rocks_row,pos_of_rocks_column+1)
                if input_lines[next_coor[0]][next_coor[1]]=='.':
                    input_lines[next_coor[0]][next_coor[1]]='O'
                    input_lines[pos_of_rocks_row][pos_of_rocks_column]='.'
                if input_lines[next_coor[0]][next_coor[1]]=='#':
                    pass
                else:
                    pass
            previous_rocks_pos = rocks_pos
    # print('done east')   
    return input_lines
    
    ####################################################################
    
```

```python
all_cycles = []

lines = input_lines
for j in range(1000000):
    lines = one_cycle(lines)
    
    if lines in all_cycles:
        i = all_cycles.index(lines) + 1
        print('found from ',i, " to ", j+1 )
        break

    all_cycles.append(lines)


remaining_cycles = (1000000000 - i) % (j+1 - i) 
print(remaining_cycles)

for k in range(remaining_cycles):
    lines = one_cycle(lines)

sum_all = 0
for index, list_char in enumerate(lines):
    sum_all += list_char.count('O')*(len(lines)-index)
sum_all
```
