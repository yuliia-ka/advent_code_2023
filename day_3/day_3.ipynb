```python
digits_all = {}

for i in range(len(input_lines)):
    digits_all[i] = [(m.start(0), m.end(0)) for m in re.finditer("\d+",input_lines[i])]
    
digits_all
```

```python
mult_valid_digits = 0


for x,y in symbol_coor2:
    
    mult_iter = []
    
    nxts=[(x-1,y),
        (x+1,y),
        (x,y-1),
        (x,y+1),
        (x-1,y+1),
        (x+1,y+1),
        (x-1,y-1),
        (x+1,y-1)]
    
    for row, digs in digits_all.items():
        for b,c in digs:
            if any((b<=j<c) and (i==row) for i,j in nxts):
                mult_iter.append(input_lines[row][b:c])

    if len(mult_iter)==2:
        mult_valid_digits += int(mult_iter[0])*int(mult_iter[1])
    
            
mult_valid_digits
```
