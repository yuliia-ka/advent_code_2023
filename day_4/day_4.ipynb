```python
#input_lines = [a.replace('Card   ','') for a in input_lines]
#input_lines = [a.replace('Card  ','') for a in input_lines]
input_lines = [a.replace('Card ','') for a in input_lines]
input_lines = [a.split(': ') for a in input_lines]
input_lines = {a:b.split(' | ') for a,b in input_lines}
input_lines
```

```python
for k,v in input_lines.items():
    #c = [v[0].replace('  ',' ').split(' '),v[1].replace('  ',' ').split(' ')]
    c = [v[0].split(' '),v[1].split(' ')]
    input_lines[k] = c

input_lines
```

```python
input_lines_copy = input_lines.copy()

wanted_keys = list(input_lines_copy.keys())[:7] # The keys you want


for k,v in dict((k, input_lines_copy[k]) for k in wanted_keys if k in input_lines_copy).items():
    #convert to int
    for i in range(len(v)):
        for j in range(len(v[i])):
            v[i][j] = int(v[i][j])

input_lines_copy
```

```python
for k,v in input_lines.items():
    input_lines[k] = [value for value in input_lines[k][0] if value in input_lines[k][1]]
input_lines
```

```python
for k,v in input_lines.items():
    input_lines[k] = len(v)
input_lines
```

```python
input_lines_n = {}

for k,v in input_lines.items():
    if v==0:
        input_lines_n[k] = 0
    else:
        input_lines_n[k] = 1*(2**(v-1))

print('Response is: ',sum(input_lines_n.values()))
```
