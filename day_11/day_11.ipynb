```python
coors_rows = []
coors_columns = []

#collect all the galaxies' coords

for idx, a in enumerate(input_lines):
    if "#" in a:
        coor_row = [idx]
        corr_col = [pos for pos, char in enumerate(a) if char == '#']
        coor_row = coor_row*len(corr_col)

        coors_rows.extend(coor_row)
        coors_columns.extend(corr_col)

len(coors_rows)
```

```python
coor_list = []

for a in range(len(coors_rows)):
    coor_list.append((coors_rows[a],coors_columns[a]))

#coor_list
```

```python
from itertools import combinations
 
pairs = list(combinations(coor_list, 2))
pairs[0][0]
```

```python
pairs[0]
```

```python
# calculate manh dist
# Manhattan Distance = | x 1 − x 2 | + | y 1 − y 2 |
dists = []
for a in pairs:
    x,y=a
    dists.append( abs(x[0]-y[0]) + abs(x[1]-y[1])  )
#sum(dists)
```

```python
mod_dists = []
for a in pairs:
    x,y=a
    dis =  abs(x[0]-y[0]) + abs(x[1]-y[1])  

    rows_in = len([r for r in empty_rows if min(x[0],y[0])<=r<=max(x[0],y[0])])
    cols_in = len([c for c in empty_columns if min(x[1],y[1])<=c<=max(x[1],y[1])])

    dis = dis + rows_in + cols_in 

    mod_dists.append(dis)


sum(mod_dists)   
```
