```python
big_thing = []
for a in input_lines:
    
    new_level = [a]
    
    level=1
    
    #while not all([x==0 for x in new_level[-1]]):
    while any([x!=0 for x in new_level[-1]]):
        new_level.append([t - s for s, t in zip(new_level[level-1], new_level[level-1][1:])])
        level += 1

    big_thing.append(new_level)
big_thing
```

```python
all_sum = 0 

for a in big_thing:
    #for b in a:
    all_sum+=sum([item[-1] for item in a])
        
all_sum
```

```python
# part 2
all_sum = 0 

for a in big_thing:
    this_element = []
    for b in a:
        this_element.append(b[0])
    this_element.reverse()

    difference_to_be_updated = 0 
    for i in this_element:
        difference_to_be_updated = i - difference_to_be_updated
    all_sum += difference_to_be_updated
        
all_sum
```
