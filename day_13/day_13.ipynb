```python
with open('input_string.txt') as f:
    input_lines = f.readlines()

input_lines = [a.replace('\n','') for a in input_lines ]
input_lines[:10]
```

```python
# list position of ' ' in input_lines
space_pos = [i for i, str in enumerate(input_lines) if str == '']
space_pos.append(len(input_lines))
space_pos
```

```python
input_lines_mod = []
beg = -1
for i in space_pos:
    chunk = [list(b) for b in input_lines[beg+1:i]]
    chunk = [[1 if a == '#' else 0 for a in b] for b in chunk]
    input_lines_mod.append(np.array(chunk))
    beg = i

input_lines_mod
```

```python
row_list = []
for i in input_lines_mod:

    rows = i.shape[0]
    cols = i.shape[1]
    #print(rows,cols)

    for row in list(range(0,(rows-1))):

        upper_half = i[0:row+1,:]
        lower_half = i[row+1:rows,:]
        size = min(len(upper_half),len(lower_half))
        upper_half = np.flipud(upper_half)[0:size,:]
        lower_half = lower_half[0:size,:]
        
        if np.array_equal(upper_half,lower_half):
            #print('Horizontal split is',row)
            row_list.append((1+row)*100)

row_list
   
        
```

```python
col_list = []

for i in input_lines_mod:

    rows = i.shape[0]
    cols = i.shape[1]
    #print(rows,cols)

    for col in list(range(0,(cols-1))):
        #print(col)

        left_half = i[:,0:col+1]
        right_half = i[:,col+1:cols]
        size = min((left_half.shape[1]),(right_half.shape[1]))
        left_half = np.fliplr(left_half)[:,0:size]
        right_half = right_half[:,0:size]
        
        if np.array_equal(left_half,right_half):
            #print('Vertical split is',col)
            col_list.append(col+1)

col_list
        
```
